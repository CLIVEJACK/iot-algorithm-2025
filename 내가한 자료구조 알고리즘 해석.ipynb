#### 1. knapsck Problem

```python
#1.
#501 p 소스코드랑 같이 보면서 풀기
def knapsack():
    array = [[0 for _ in range(maxWeight+1)] for _ in range(rowCount + 1)] # 2차원 배열 5x8
    for row in range(1, rowCount +1):
        for col in range(1, maxWeight+1):
            if weight[row] > col : # 이전값보다 컬럼(무게)가 작으면
                array [row][col] = array[row-1][col] # 이전값을 그대로 가지고 내려오고
            else:
                value1 = money[row] + array[row-1][col-weight[row]] # 최대무게에서 현재들어갈 무게를 뺀 위치의 가치+ 현재보석의 가격
                value2 =  array[row-1][col] # 이전위치 값
                array[row][col] = max(value1, value2) # 둘중에 비싼걸
    return array[rowCount][maxWeight] # 최종 행렬 마지막값(array[4][7]) 을 리턴

maxWeight = 7
rowCount = 4
weight = [0,6,4,3,5]
money = [0,13,8,6,12]

maxValue = knapsack()
print(f'배낭 = {maxValue}')
```

```python
# 2번
def isQueueFull():
    global SIZE, queue, front, rear
    if rear != SIZE - 1:
        return False
    elif rear == SIZE - 1 and front == -1 :
        return True
    else:                 # p253 페이지 보기 
        for i in range(front+1 , SIZE):
            queue[i-1] = queue[i] # 데이터앞 빈자리에 첫번째 데이터 옮김
            queue[i] = None

        front -= 1
        rear -= 1
        return False

    
def isQueueEmpty():
    global SIZE, queue, front, rear
    if front == rear:
        return True
    else:
        return False
    
def enQueu(data): # enQueue 에 집어넣을 싸이즈(data)를 넣음
    global SIZE, queue, front, rear
    if isQueueFull():
        print('Queue is Full')
    else:
        rear += 1
        queue[rear] = data

def deQueue(): # deQueue 는 빼는거 
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('Queue is empty')
        return None  # 리턴 None 왜 한지 보기 
    else: # 이 로직 뭔지 보기 
        front += 1
        data = queue[front]
        queue[front] = None
        return data
    
def peek():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('Queue is empty')
        return None
    else:
        return queue[front+1]



# 1.초기화
SIZE = 5
queue = [None for _ in range(SIZE)]
front = rear = -1

## 3.메인 모듈
if __name__=='__main__':
    while True:
        select = input('삽입(I)/추출(E)/확인(V)/종료(Q) -->').upper()

        if select == 'Q':
            break
        elif select == 'I':                                                 # while Ture 써서 바꾸기 
            data = input('데이터 입력 > ')  
            enQueu(data) # enQueue는 엔터 큐 데이터 넣기 
            print(f'대기줄 상태 : {queue}')
        elif select == 'E':
            data =deQueue()
            print(f'손닙입장 : {data}')
            print(f'대기줄 상태: {queue}')
        elif select == 'V':
            data = peek()
            print(f'대기자명 : {data}')
            print(f'대기자 상태 : {queue}')

        else:
            print('입력 오류')
```
